Programa iterativo que calcula el n termino de la  la serie de Fibonacci

```python
import numpy as np
import math
from matplotlib import pyplot
import time
```

```python
def fibonacci(n):
    a = 0
    b = 1
    for i in range(n):
        r = a+b
        b = a
        a = r
    return a
```

```python
X = []
Y = []
n = 100
for i in range (n):
    start_time = time.time()
    fibonacci(i)
    X.append(i)
    Y.append(time.time() - start_time)
```

```python
pyplot.plot([X[j] for j in range(len(Y))], [Y[j] for j in range(len(X))],color="green")
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.xlim(0,101)
pyplot.ylim(-0.000001, 0.003)
pyplot.title("Fibonacci")
pyplot.xlabel("Iteracion")
pyplot.ylabel("Valor")
pyplot.show()
```
