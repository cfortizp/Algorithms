Este codigo genera la distribucion y el numero de comparaciones de Insertion Sort

```python
import itertools
import numpy as np
import math
from matplotlib import pyplot
import random

def randomPerm(n):
    v=[]
    for i in range(n):
        v.append(i+1)
    for i in range(len(v)-1):
        j = random.randint(i, len(v)-1)
        aux = v[i]
        v[i] = v[j]
        v[j] = aux
    return v

def permutations(n):
    i = 0
    permutaciones = []
    while i < math.factorial(n):
        lista = randomPerm(n)
        if not lista in permutaciones:
            permutaciones.append(lista)
            i+=1
    permutaciones.sort()
    return(permutaciones)

def comp_insertion_sort(A):
    comparaciones = 0
    i = 1
    while i < len(A):
        j = i-1
        while (j >= 0) and (A[j+1] < A[j]):
            comparaciones+=1
            element = A[j]
            A[j] = A[j+1]
            A[j+1] = element
            j = j-1
        i = i+1
        comparaciones+=1
    return(comparaciones)
  

A = [4,3,2,1]
j = 0
X = []
Y = []
Z = []
suma = 0;
for i in itertools.permutations(A,4):
    X.append(j)
    Y.append(comp_insertion_sort(list(i)))
    suma += Y[j]
    j +=1

j = 0
suma1 = 0
for i in permutations(4):
    Z.append(comp_insertion_sort(i))
    suma1 += Z[j]
    j +=1
  
  

print ("Itertools")
print ("El promedio de numero de comparaciones es: ",suma/24)
print ("Permutacion Recursiva")
print ("El promedio de numero de comparaciones es: ",suma1/24)
pyplot.plot([X[j] for j in range(len(Y))], [Y[j] for j in range(len(X))],color="green")
pyplot.plot([X[j] for j in range(len(Y))], [Z[j] for j in range(len(X))],color="red")
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
pyplot.xlim(0,j)
pyplot.ylim(0, max(Y)+1)
pyplot.title("Comparaciones")
pyplot.xlabel("Iteracion")
pyplot.ylabel("Valor")
pyplot.show()
```
